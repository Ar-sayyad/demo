```python
jc = df[(df['Description'] == 'Joined A Contest')]
jccnt = sum(jc['Amount'])

wc = df[(df['Description'] == 'Won A Contest')]
wccnt = sum(wc['Amount'])

wdr = df[(df['Description'] == 'Withdrew Cash')]
wdrs = sum(wdr['Amount'])

dc = df[(df['Description'] == 'Deposited Cash')]
dccnt = sum(dc['Amount'])

pending = (wccnt - dccnt)
print("Total Joined:",jccnt," Total Won:",wccnt," Total Deposit:", dccnt," Withdraw Cash:",wdrs, " Pending:",pending)
data = {
    'Description': ['Total Joined','Total Won','Total Deposit','Withdraw Cash','Pending'],    
    'Amount': [jccnt,wccnt,dccnt,wdrs,pending]
}
dff = pd.DataFrame({
                    'Description': ['Total Joined','Total Won','Total Deposit','Withdraw Cash','Pending'], 
                    'Amount': [jccnt,wccnt,dccnt,wdrs,pending]
                    },
                    index=['Joined','Won','Deposit','Withdraw','Pending'])
dff
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
labels = 'Joined','Won','Deposit','Withdraw','Pending'
sizes =  [jccnt,wccnt,dccnt,wdrs,pending]
explode = (0.05,0.05,0.05,0.05,0.05)
colors = ['#ff6666', '#ffcc99', '#99ff99', '#66b3ff', '#c2c2f0','#ffb3e6', '#c2c2f0','#ffb3e6']
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.1f}% \n (Rs.{v:d})'.format(p=pct,v=val)
    return my_autopct

fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors, explode=explode,  pctdistance=0.85,  labels=labels, autopct=make_autopct(sizes),
        shadow=False,startangle=90)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
ax1.axis('equal')
plt.tight_layout()
plt.show()
```
