```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GOOGL','AMZN','MSFT','FB']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.bar(company,revenue,width=0.5)
```

```python
plt.xticks(ypos,company)
#plt.xlabel("Company")
plt.ylabel("Revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(ypos-0.2,revenue,width=0.4, label="Revenue")
plt.bar(ypos+0.2,profit,width=0.4, label="Profit")
plt.legend()
plt.grid(color="gray",linestyle="-.",linewidth=0.9)
```

```python
plt.yticks(ypos,company)
plt.title("US Tech Stocks")
plt.xlabel("Revenue(bln)")
plt.barh(ypos-0.2,revenue,height=0.4, label="Revenue")
plt.barh(ypos+0.2,profit,height=0.4, label="Profit")
plt.legend()
plt.grid(color="gray",linestyle="dashed")
```
