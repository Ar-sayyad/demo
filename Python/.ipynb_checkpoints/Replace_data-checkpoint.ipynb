```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather.csv")
df
```

```python
ndf = df.replace(-99999,np.NaN)
ndf
```

```python
ndf = df.replace([-99999,-88888],np.NaN)
ndf
```

```python
ndf = df.replace([-99999,-88888,'0'],np.NaN)
ndf
```

```python
ndf = df.replace({
    'Temperature': -99999,
    'Windspeed':-99999,
    'Events': '0'
},np.NaN)
ndf
```

```python
ndf = df.replace({
    -99999: np.NaN,
    'No Event': 'Sunny'
})
ndf
```

```python
ndf = df.replace('[A-Za-z]','',regex=True)
ndf
```

```python
ndf = df.replace({
    'Temperature': '[A-Za-z]',
    'Windspeed': '[A-Za-z]'
},'',regex=True)
ndf
```
