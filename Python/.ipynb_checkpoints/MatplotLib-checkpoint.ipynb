```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,41,52,40,47,50,42]
```

```python
plt.plot(x,y,color='green',linewidth=3, linestyle='dotted')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth=3,linestyle='dashed')
```

```python
plt.plot(x,y,'g+--')
```

```python
plt.plot(x,y,'bD-.')
```

```python
plt.plot(x,y,color='#008040',marker='D',linestyle='dotted')
```

```python
plt.plot(x,y,color='#ff0000',marker='D',linestyle='',markersize=20)
```

```python
plt.plot(x,y,color='green',marker='*',linestyle='',markersize=20)
```

```python
plt.plot(x,y,color='green',alpha=0.5)
```
